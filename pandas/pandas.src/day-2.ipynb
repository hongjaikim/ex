## 4-4 여러개의 데이터프레임을 하나의 엑셀파일로 저장

```python
data1 = {'name':['Jerry','Riah','Paul'],
        'algol' : ['A','A+','B'],
        'basic' : ['C','B','A'],
        'c++' : ['B+','A','C']}
data2 = {'c0' : [1,2,3],
        'c1': [4,5,6],
        'c2' : [7,8,9],
        'c3' : [10,11,12],
        'c4' : [13,14,15]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# df1의 name을 인덱스 컬럼으로
# df2의 c0를 인덱스 컬럼으로
df1.set_index('name',inplace=True)
df2.set_index('c0',inplace=True)

```

```python
# df1을 엑셀의 'sheet1'로, df2를 'sheet2'로 df_excel.xlsx로 파일생성
# 1. file을 ExcelWriter로 포인터지정 : open
# 2. sheet를 저장
# 3. file.save() 파일 저장   :close
file = pd.ExcelWriter('./dataset/df_excel.xlsx')
df1.to_excel(file, sheet_name='sheet1')
df2.to_excel(file, sheet_name='sheet2')
file.save()
```
