```python
import pandas as pd
import seaborn as sns
# seaborn에서 titanic 데이터를 로드하여 데이터프레임 titanic 으로 생성
titanic = sns.load_dataset('titanic')
# titanic 컬럼별 개수 확인
print(titanic.count())
print()
# titanic 전체 기본 정보 확인
print(titanic.info())
print()
# titanic 통계 정보 확인
print(titanic.describe())
print()
# 컬럼별 개수가 300 미만인 컬럼과 데이터 타입이 숫자가 아닌 컬럼은 
# 제외하고 나머지 컬럼만 추출하여 데이터프레임 (df) 생성
cnt = titanic.count()          #컬럼별 데이터 개수-> 시리즈
col_list = cnt[cnt>=300].index   # 데이터 개수가 300 이상 컬럼명 추출
df = titanic.loc[:,col_list]
(df.dtypes == 'int64') | (df.dtypes == 'float64')
df = df.loc[:,(df.dtypes == 'int64') | (df.dtypes == 'float64')]
print(df)
# df 의 기본정보
print(df.info())
# df 처음 5개의 정보만 
print(df.head())
# df의 age와 fare컬럼에서 200번째 부터 400번째 자료까지만 추출하여 df_1에 저장
df_1 = df.loc[200:400,['fare','age']]
# df_1의 age컬럼을 index 컬럼으로 
df_1.set_index('age',inplace=True)

```
