```python
# 모듈을 프로그램에 삽입 import 모듈명 -> 모듈명.함수 또는 모듈명.변수
import math

print(math.ceil(43.2))          # 올림 정수
print(math.floor(43.2))         # 내림 정수
print(math.sin(1))


```

```python
# 모듈명이 길어서 약어로 축약해서 사용하고자 할 경우
import math as m
print(m.ceil(43.2))          # 올림 정수
print(m.floor(43.2))         # 내림 정수
print(m.sin(30))
```

```python
from math import sin, cos, floor, ceil

print(ceil(43.2))          # 올림 정수
print(floor(43.2))         # 내림 정수
print(sin(30))
```

```python
import random
print("# random 모듈")

# 0.0~1.0 사이 float 리턴
print("- random(): ",random.random())

# uniform(min,max): 지정한 범위 사이의 float를 리턴합니다
print("- uniform(10,20): ",random.uniform(10,20))

# randrange() : 지정한 범위의 int 리턴
# - randrange(max) : 0부터 max 사이의 값 리턴
# - randrange(min, max) : min부터 max 사이의 값 리턴
print("- randrange(10): ",random.randrange(10))

# choice(list): 리스트 내부에 있는 요소를 랜덤하게 선택합니다.
print("- choice([1,2,3,4,5]): ",random.choice([1,2,3,4,5]))

# shuffle(list) : 리스트의 요소들을 랜덤하게 섞습니다.
print("- shuffle([1,2,3,4,5]): ",random.shuffle([1,2,3,4,5]))

# sample(list, k=<숫자>): 리스트의 요소 중에 k개를 뽑습니다
print("- sample([1,2,3,4,5],k=2): ",random.sample([1,2,3,4,5,6,7,8,9,0],k=4))


```
